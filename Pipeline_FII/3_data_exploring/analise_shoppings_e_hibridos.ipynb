```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/2_data_clean/data_cleanned_20241115_105609.csv", sep=";")
```

```python
data
```

```python
data_shops_hib = data[(data['SEGMENTO'] == 'Híbrido') | (data['SEGMENTO'] == 'Shoppings')]
data_shops_hib
```

```python
data_shops_hib['LIQUIDEZ'].describe()
```

```python
# Cria o box plot horizontal 
plt.figure(figsize=(10, 6)) 
sns.boxplot(x='LIQUIDEZ',data=data_shops_hib, orient='h') 
plt.title('Box Plot por Categoria') 
plt.xlabel('Valor') 
plt.ylabel('Categoria') 
plt.show()
```

```python
df_filter = data_shops_hib[(data_shops_hib['LIQUIDEZ'] >= 3.887500e+04) & (data_shops_hib['LIQUIDEZ'] <= 5.255568e+05)]
df_filter
```

```python
df_filter.describe()
```

```python
sns.boxplot(x='DIVIDEND_YIELD', data=df_filter, orient='h')
plt.show()
```
